```python
#importer la bibilotque

import networkx as nx
import matplotlib.pyplot as plt
```

```python
#la création du graphe vide
G = nx.Graph()
```

```python
### Ajouter des nœuds

G.add_node(1)
G.add_nodes_from([2,3])

# Afficher path_graph 
H = nx.path_graph(10)
plt.figure(figsize=(10,7))
nx.draw ( H , with_labels = True ,font_weight='bold')

# ajouter les noeuds de H dans G
G.add_nodes_from(H)

# ajouter H comme noeud de G
G.add_node(H)
```

```python
## Afficher le graph G
plt.figure(figsize=(10,7))
nx.draw ( G , with_labels = True ,font_weight='bold')
```

```python
## Ajouter des arêtes

G.add_edge(1,2)
G.add_edge(1,H)
Tuple = (0,9)
G.add_edge(*Tuple)

G.add_edges_from(H.edges)
plt.figure(figsize=(10,7))
nx.draw(G, with_labels = True)
```

```python
## Exploration du graphe

G.add_node("spam")        # adds node "spam"
G.add_nodes_from("spam")  # adds 4 nodes: 's', 'p', 'a', 'm'
G.add_edge(3, 'm')

plt.figure(figsize=(10,7))
nx.draw(G ,with_labels = True)




 
```

```python
##########################################
#####__________APPLICATION___________#####
##########################################

###créer le graph:
DG = nx.DiGraph()

# definition des noeuds
DG.add_node(1,label='1',color='red')
DG.add_node(2,label='2',color='red')
DG.add_node(3,label='3',color='white')
DG.add_node(4,label='4',color='white')
DG.add_node(5,label='5',color='red')
DG.add_node(6,label='6',color='red')
# definition des aretes
DG.add_edge(1,2,weight=2.0,style='dashed')
DG.add_edge(1,3,weight=4.0,style='solid')
DG.add_edge(3,5,weight=3.0,style='solid')
DG.add_edge(2,3,weight=1.0,style='solid')
DG.add_edge(5,6,weight=2.0,style='dashed')
DG.add_edge(4,6,weight=2.0,style='solid')
DG.add_edge(5,4,weight=3.0,style='solid')
DG.add_edge(2,4,weight=4.0,style='solid')
DG.add_edge(2,5,weight=2.0,style='dashed')

pos = nx.shell_layout(DG)
weights = nx.get_edge_attributes(DG,'weight')

plt.figure(figsize=(10,7))
nx.draw(DG, with_labels = True, pos = pos,node_size = 1500 ,node_color = 'r',edge_color = 'g',arrowsize = 33, font_size =16)
nx.draw_networkx_edge_labels(DG , pos , edge_labels = weights , font_size =16)
```

```python
#appliquer djikstra :

length,path=nx.bidirectional_dijkstra(DG,1,6)
path_edges = ((1,2),(2,5),(5,6))
print(length)
print(path)
```

```python
plt.figure(figsize=(10,7))
color_nodes = ['red' if i not in path else 'grey' for i in DG.nodes ]
color_edges = ['green' if j not in path_edges else 'red' for j in DG.edges ]
nx.draw(DG, with_labels = True, pos = pos, node_color = color_nodes, edge_color = color_edges ,node_size = 1000 )
nx.draw_networkx_edge_labels(DG , pos , edge_labels = weights )
```

```python
#appliquer le plus court chemin :
print (nx.shortest_path(DG, 1, 6))
```

```python
#liste des arrets:
nx.to_pandas_edgelist(DG)
```

```python
#matrice d'adjacence:
nx.to_pandas_adjacency(DG)
```

```python
# créer path_edges
length,path=nx.bidirectional_dijkstra(DG,1,6)
path_edges = ((1,2),(2,5),(5,6))
g=nx.DiGraph()
g.add_nodes_from(path)
edges =[(path[0],path[1])]
if len(path)>=2:   
    for i in range(1,len(path)):
        for j in range(i,len(path)):
            if i!=j:
                edges.append((path[i],path[j]))
            else:
                continue
            
g.add_edges_from(edges)
g.edges
```
